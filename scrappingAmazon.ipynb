```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# The webpage URL
URL = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_"

soupArr = []
# HTTP Request
pages = 20
for i in range(1, pages, 1):
    find = []
    while len(find) == 0:
        webpage = requests.get(URL + str(i))
        soup = BeautifulSoup(webpage.content, "html.parser")
        find = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    soupArr.append(soup)
# Soup Object containing all data
```

```python
productUrl = []
productName = []
productPrice = []
rating = []
review = []
for soup in soupArr:
    ele = soup.find_all("div", attrs={'class': 'a-section a-spacing-small puis-padding-left-small puis-padding-right-small'})
    # # Remove the class from the element
    # print(ele)
    for i in ele:
        i.extract()
    # Fetch links as List of Tag Objects
    find = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    # Store the links
    # Loop for extracting links from Tag Objects
    for element in find:
            productUrl.append("https://www.amazon.com" + element.get('href'))
    print(productUrl)

    find = soup.find_all("span", attrs={'class':'a-size-medium a-color-base a-text-normal'})
  
    for element in find:
            productName.append(element.text.strip())
    print(productName)

    find = soup.find_all("span", attrs={'class':'a-price-whole'})

    for element in find:
            productPrice.append(element.text.strip())
    print(productPrice)

    find = soup.find_all("span", attrs={'class':'a-icon-alt'})
    while len(productPrice) != len(productName):
        productPrice.pop();
        
    for element in find:
            if element.text.strip()[2] == 'S':
                continue
            rating.append(element.text.strip())
    #         print(element.text.strip())
    print(rating)

    find = soup.find_all("span", attrs={'class':'a-size-base s-underline-text'})
 
    for element in find:
            review.append(element.text.strip())
    print(review)
    
    print(len(productUrl), len(productName), len(productPrice), len(rating), len(review))
```

```python
d = { "productName":productName, "productUrl":productUrl, "price": productPrice, "rating": rating, "review":review}
amazon_df = pd.DataFrame.from_dict(d)
amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
